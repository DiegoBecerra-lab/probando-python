# **Leyendo archivos CSV**

## **Importando archivos CSV**

```python
import pandas as pd
```

```python
url='/content/superstore_data.csv'
datos=pd.read_csv(url)
```

```python
datos.head()
```

```python
url_punto_coma='/content/superstore_data_punto_coma.csv'
datos_punto_coma=pd.read_csv(url_punto_coma, sep=';')
```

```python
datos_punto_coma.head()
```

## **Parámetros read_csv**

```python
#nrows=Para leer n números de filas dentro del archivo

datos_primeras_lineas=pd.read_csv(url, nrows=5)
datos_primeras_lineas
```

```python
#usecols=Para leer solamente algunas columnas seleccionadas

datos_seleccion=pd.read_csv(url, usecols=['Id','Year_Birth','Income'])
datos_seleccion
```

```python
datos_seleccion=pd.read_csv(url, usecols=[0,1,4])
datos_seleccion
```

## **Exportando archivos CSV**

```python
#index=Genera nuevos índices

datos_seleccion.to_csv('clientes_mercado.csv')
```

```python
clientes_mercado=pd.read_csv('clientes_mercado.csv')
clientes_mercado
```

```python
datos_seleccion.to_csv('datos_mercado.csv', index=False)
```

# **Leyendo archivos Excel**

```python
url='/content/emisiones_CO2.xlsx'

import pandas as pd
datos_co2=pd.read_excel(url)
```

```python
datos_co2.head()
```

## **Parámetros read_excel**

```python
#sheet_name: Para leer una hoja de excel en específico

percapita=pd.read_excel(url, sheet_name='emisiones_percapita')
```

```python
percapita.sample(5)
```

```python
fuentes=pd.read_excel(url, sheet_name='fuentes')
```

```python
fuentes.tail()
```

```python
intervalo=pd.read_excel(url, sheet_name='emisiones_C02', usecols='A:D')
```

```python
intervalo
```

```python
intervalo_2=pd.read_excel(url, sheet_name='emisiones_C02', usecols='A:D', nrows=10)
```
